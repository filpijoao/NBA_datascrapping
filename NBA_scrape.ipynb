### Imports básicos

Primeiro o download o site e a criação do objeto

```python
url_homepage = "https://www.basketball-reference.com/"
```

```python
response_homepage = requests.get(url_homepage)
```

Criação de uma lista vazia

```python
teams_href_list =[]
```

Extração dos time do Leste

```python
east_data = soup_homepage.find('table', id='confs_standings_E').find_all('tr')
```

```python
for entry in east_data[1:]:
    teams_href_list.append(entry.find_all("a")[1].attrs['href'])
```

Extração dos time do Oeste

```python
west_data = soup_homepage.find('table', id='confs_standings_W').find_all('tr')
```

```python
for entry in west_data[1:]:
    teams_href_list.append(entry.find_all("a")[1].attrs['href'])
```

# Criando Lista com todos as temporadas de cada time 

Criando lista vazia

```python
seasons_href_list = []
```

Criando função para extrair as temporadas de cada time

```python
def get_urls(team_url, lists_):
    
    #Definindo o nome do time e criando URL
    team = team_url[7:-1]
    url_teams = f"https://www.basketball-reference.com{team_url}"
    
    #Fazendo Download od site e criando o meu objeto
    response_teams = requests.get(url_teams)
    soup_teams = BeautifulSoup(response_teams.content, "html.parser")
    
    #Achando a minha tabela de download
    soup_teams_data = soup_teams.find('table', id=team).find_all('tr')
    
    #Extraindo as urls e adicionado a lista
    for entry in soup_teams_data[1:]:
        lists_.append(entry.find('a').attrs['href'])
```

Fazendo o Download do link de todas as temporadas

```python
for entry in teams_href_list:
    get_urls(entry, seasons_href_list)
    print(f"finishing {entry[7:-1]}")
```

# Selecionando apenas as temporadas desejadas

```python
finals_seasons_href_list = []
```

```python
for entry in seasons_href_list:
    num = int(entry[11:15])
    if num > 1979 and num != 2025:
        finals_seasons_href_list.append(entry)
```

```python
finals_seasons_href_list
```

# Criando dicionário de estatíticas

Criando URL genérica para coletar as chaves do dicionário

```python
url = f'https://www.basketball-reference.com{finals_seasons_href_list[1]}'
```

```python
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
dict_names = soup.find('table', id='totals').find('tr').find_all('th')
```

```python
stats_dict ={
    "Team" : [],
    "Season" : []
}
for i in dict_names[1:]:
    stats_dict[i.text] = []
```

# Criando função para extrair estatíticas

```python
def get_stats(season_url, dict_):
    
    #Defininado time, anos e url
    team = season_url[7:10]
    year = season_url[11:15]
    url_season = f"https://www.basketball-reference.com{season_url}"
    
    #Fazendo Download do site e criando o meu objeto
    response_season = requests.get(url_season)
    soup_season = BeautifulSoup(response_season.content, "html.parser")
    
    #Achando a tabela desejada
    total_data = soup_season.find('table', id='totals').find_all('tr')
    
    #Fazendo o Download dos dados
    for entry in total_data[1:-1]:
        player_stats = entry.find_all('td')
        dict_entry = list(dict_.keys())
        dict_['Team'].append(team)
        dict_['Season'].append(year)
        for i, j in zip (dict_entry[2:], player_stats):
            dict_[i].append(j.text)
```

# Fazendo o download das estatíticas 

```python
for count, entry in enumerate(finals_seasons_href_list):
    
    get_stats(entry, stats_dict)
    
    if count % 10 == 0:
        print (f"{count} temporadas baixadas")
```

```python
df = pd.DataFrame(stats_dict)
```

```python
df = df.drop_duplicates()
```
