### Importes Básicos

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

Criação do Dataframe

```python
df = pd.read_csv('NBA_database.csv', index_col=0)
df
```

Elminação de Na

```python
df = df.fillna(0)
```

# Excluindo as temporadas do locaute

```python
df.groupby('Season').sum()['GS']/30
```

```python
df.describe()
```

```python
df_numeric = df.iloc[:,7:]
```

```python
# Configuração das subplots
fig, axs = plt.subplots(nrows=int(len(df_numeric.columns)/2), ncols=2, figsize=(10, 30))

# Loop para criar os gráficos
for n, col in enumerate(df_numeric.columns):
    # Converter o índice 'n' em índices de linha e coluna
    row = n // 2
    col_idx = n % 2
    df_numeric[col].(ax=axs[row, col_idx])  # Corrigir a referência ao 'ax' correto
    axs[row, col_idx].set_title(col)  # Adiciona o nome da coluna como título de cada gráfico

plt.tight_layout()
plt.show()
```
